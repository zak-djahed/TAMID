```python
import numpy as np
import pandas as pd
import yfinance as yf
import random

# Fetch S&P 500 tickers
tickers_df = pd.read_html("https://en.wikipedia.org/wiki/List_of_S%26P_500_companies")[0]
all_tickers = tickers_df['Symbol'].tolist()

# Select 200 random tickers
random_tickers = random.sample(all_tickers, 200)

# Fetch stock price data (1 month of daily data)
stocks_data = yf.download(random_tickers, period="1mo", interval="1d")['Adj Close']

# Fetch market index data (S&P 500)
market_data = yf.download("^GSPC", period="1mo", interval="1d")['Adj Close']

# Compute daily returns for stocks and market
stocks_returns = stocks_data.pct_change().dropna()
market_returns = market_data.pct_change().dropna()

# Align dates (ensure stocks and market have the same timestamps)
# Ensure both indexes are timezone-naive before aligning
stocks_returns.index = stocks_returns.index.tz_localize(None)
market_returns.index = market_returns.index.tz_localize(None)

# Now align dates
stocks_returns, market_returns = stocks_returns.align(market_returns, join='inner', axis=0)


# Compute betas using NumPy
betas = {}

for ticker in stocks_returns.columns:
    beta, alpha = np.polyfit(market_returns, stocks_returns[ticker], 1)  # Perform linear regression
    betas[ticker] = beta  # Store Beta coefficient

# Convert betas into a DataFrame and rank them
betas_df = pd.DataFrame.from_dict(betas, orient='index', columns=['Beta']).sort_values(by='Beta', ascending=False)

# Create portfolios based on beta rankings
top_20_high_beta = betas_df.head(20)  # High beta stocks
top_20_low_beta = betas_df.tail(20)   # Low beta stocks

# Display results
# Display results using Pandas print or .to_csv()
print("Market Betas:\n", betas_df)
print("\nHigh Beta Portfolio:\n", top_20_high_beta)
print("\nLow Beta Portfolio:\n", top_20_low_beta)

# Save to a CSV file
betas_df.to_csv("market_betas.csv")
top_20_high_beta.to_csv("high_beta_portfolio.csv")
top_20_low_beta.to_csv("low_beta_portfolio.csv")
```
